```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
bank=pd.read_csv("/content/Customer-Churn-Records.csv")
```

```python
bank.head(10)
```

```python
le=LabelEncoder()
le2=LabelEncoder()
```

```python
bank['new_geo']=le.fit_transform(bank['Geography'])
bank['new_gen']=le2.fit_transform(bank['Gender'])
```

```python
ind=bank[['new_gen','new_geo','Age','Balance']]
dep=bank['Exited']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()

scaled=scaler.fit_transform(ind)
```

```python
lr=LogisticRegression()
lr.fit(scaled,dep)
```

```python
ans=lr.predict([[1,0,59,10094400.373]])

if(ans==1):
 print("Customer will exit")
else:
 print("Customer stays")
```
